```python
import numpy as np
from scipy.interpolate import interp1d
import pandas as pd
```

```python
weather = pd.read_csv("../weather/data.csv")
weather.columns
```

```python
v_ms = np.arange(1, 26)
co_vals = np.array([0, 0.1, 0.27, 0.36, 0.42, 0.46, 0.48, 0.5, 0.5, 0.5, 0.49, 0.45, 0.39, 0.34, 0.28, 0.23, 0.19, 0.16,
                    0.14, 0.12, 0.1, 0.09, 0.08, 0.07, 0.06])
```

```python
co_interp = interp1d(v_ms, co_vals, kind="linear", bounds_error=False, fill_value=0.0)
v_site = weather['WIND_SPEED'] / 3.6 #km/h to m/s
#only take values that are in the interpolated range
mask = (v_site <= 25)
co_site = co_interp(v_site[mask])
```

```python
print(f"mean: {co_site.mean():.4f}")
print(f"std deviation: {co_site.std(ddof=1):.4f}")
```
